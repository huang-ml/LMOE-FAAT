```python
import sys
import os
sys.path.append('.')
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

from modeling_qwen3_lmoe import Qwen3MoeForCausalLMConvert, Qwen3MoeForCausalLMLoad, save_moe_model
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-0.6B")
moe_model = Qwen3MoeForCausalLMConvert(
        "Qwen/Qwen3-0.6B",
        use_attention_experts=False,
        use_ffn_experts=True,
        num_attention_experts=4,
        num_attention_experts_per_token=2,
        num_ffn_experts=16,
        num_ffn_experts_per_token=8,
        rank=32,
        alpha=64,
        dropout=0.05
    )
```

```python
save_path = "./qwen3-0.6b-moe"
save_moe_model(moe_model, save_path)
tokenizer.save_pretrained(save_path)
```
